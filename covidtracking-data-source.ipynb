# covoidtracking data source

```python
import json
import urllib.request
import pandas as pd
pd.set_option('max_columns', 100)
pd.set_option('max_rows', 100)
```

```python
url_states_daily = 'https://covidtracking.com/api/states/daily'
test_daily = urllib.request.urlopen(url_states_daily).read().decode()
data_json = json.loads(test_daily)
data_csv = pd.read_csv(url_states_daily + '.csv')
```

```python
data_csv = data_csv[data_csv.state=='GU']
data_csv = data_csv.sort_values('date')
data_csv
```
